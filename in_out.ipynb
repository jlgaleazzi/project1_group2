```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
data_intake = pd.read_csv("aac_intakes.csv")
data_intake.head()
```

```python
# get outcome
data_outcome = pd.read_csv("aac_outcomes.csv")
data_outcome.head()
```

```python
# merge datasets
in_out_data = pd.merge(data_intake, data_outcome, on='animal_id')
in_out_data.head()
```

```python
# clean up 
clean_in_out_data = in_out_data.drop(columns=['animal_type_y','breed_y','color_y','datetime_y','name_y'])
clean_in_out_data.head()
```

```python
clean_in_out_data.rename(columns={'animal_type_x': 'animal_type','color_x':'color','breed_x': 'breed', 'datetime_x':'datetime','name_x':'name'}, inplace=True)
clean_in_out_data.head()
```

```python
#intake condition sick - outcome type
condition = clean_in_out_data.groupby(['intake_condition','outcome_type']).count()
conditiondf = pd.DataFrame(condition)
conditiondf
```

```python
# aged
aged = clean_in_out_data.loc[clean_in_out_data['intake_condition'] == 'Aged'] 
outcomeTypes = aged['outcome_type'].unique()
outComeGroups = []
outComeData = []
aged_outcomeTypes_total = aged['outcome_type'].count()

for otype in outcomeTypes:
    type_count = aged.loc[aged['outcome_type'] == otype]['animal_id'].count()
    avg = type_count/aged_outcomeTypes_total
    outComeGroups.append(otype) 
    outComeData.append(avg)
    #print(aged.loc[aged['outcome_type'] == otype]['animal_id'].count())

# create pie chart
colors=["salmon","gold","skyblue"]
explode = (0.1,0,0,0,0,0)
plt.title("% Outcome for Aged Animals")
plt.pie(outComeData,labels=outComeGroups, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=-80)
plt.savefig('aged')
```
